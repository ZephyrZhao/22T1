```python
# first drawing
import matplotlib.pyplot as plt
import numpy as np
import math
'''
np.arange(a)      返回一个起点是0 终点是a 步长为1的 np.array 对象
np.arange(a,b)    返回一个起点是a 终点是b 步长为1的 np.array 对象
np.arange(a,b,c)  返回一个起点是a 终点是b 步长为c的 np.array 对象
'''
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')

plt.show()
```
